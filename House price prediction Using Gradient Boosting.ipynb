# Lets Start Feature Engineering

```python
df.head()
```

```python
df = df.drop('id',axis = 1)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['date']
```

```python
#date.year is inbuilt function to separte year from date
def year_extraction(date):
    return date.year
```

```python
#apply above function using lambda expression

df['year'] = df['date'].apply(lambda date : date.year)
df['month'] = df['date'].apply(lambda date : date.month)
```

```python
df.head()
```

```python
df.groupby('month').mean()['price'].plot()
```

```python
df.groupby('year').mean()['price'].plot()
```

```python
df = df.drop('date',axis =1)
```

```python
df['zipcode'].value_counts()
```

```python
df = df.drop('zipcode',axis = 1)
```

```python
df['sqft_basement'].value_counts()
```

```python
df
```

```python
X   = df.drop('price',axis = 1).values
```

```python
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
reg = GradientBoostingRegressor()
```

# Grid SearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {'max_depth':[1,2,3,4,5,6,7,8],'n_estimators': [5,10,15,20,50,100],'learning_rate':[0.3,0.4,0.5,0.6,0.7,0.8,0.9,1,2]}
```

```python
grid = GridSearchCV(estimator=reg,param_grid=parameter,cv=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
Model = grid.best_estimator_
```

```python
prediction  = Model.predict(X_test)
```

```python
plt.figure(figsize=(15,10))

font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 20,
        }
plt.scatter(y_test,prediction)
plt.xlabel('Actal_data',fontdict=font)
plt.ylabel('prediction',fontdict=font)
plt.plot(y_test,y_test,'r--')
```

```python
from sklearn.metrics import mean_squared_error
```

# Rmse

```python
np.sqrt(mean_squared_error(y_test,prediction))
```
